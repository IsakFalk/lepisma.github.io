```python
%matplotlib inline
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

def g(x, y):
    return np.sin(x) ** 3 + np.cos(10 + y * y) * np.cos(x) ** 3
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
pop = 20
popx = np.random.rand(pop) + 1
popy = np.random.rand(pop) + 2.5
plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, 20, cmap='viridis');
plt.colorbar()
#plt.scatter(popx, popy, c="DodgerBlue")
plt.savefig("./landscape.pdf")
```

```python
pop = 20
popx = np.random.rand(pop) + 1
popy = np.random.rand(pop) + 2.5
plt.figure(figsize=(10, 6))
plt.contour(X, Y, g(X, Y), 20, cmap='viridis');
plt.colorbar()
#plt.scatter(popx, popy, c="DodgerBlue")
plt.savefig("./landscape-two.pdf")
```

```python
from scipy.stats import norm

plt.figure(figsize=(10, 6))
x_axis = np.arange(-10, 20, 0.001)
plt.plot(x_axis, norm.pdf(x_axis,0,2))
plt.plot(x_axis, norm.pdf(x_axis,10,2), linestyle="dashed")
plt.arrow(2.5, 0.175, 5, 0.00, head_width=0.01, head_length=0.4, fc='gray', ec='gray')
plt.text(5, 0.16, "ε", name="Source Sans Pro", size=20)
plt.xticks([])
plt.yticks([])
plt.savefig("./drift.pdf")
```
